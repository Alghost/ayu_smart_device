```python
import pandas as pd
import numpy as np

ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()

ts.plot()
```

```python
df = pd.read_csv('bike.csv')
df_sub = df['count']
df_ts = df_sub.cumsum()

df_ts.plot()
df
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
df.plot()
```

```python
df = pd.read_csv('bike.csv')
df_sub = df.loc[:, "casual":"count"]
df_sub.index = df['datetime']

df_ts = df_sub.cumsum()

df_ts.plot()
```

```python
df = pd.read_csv('bike.csv')
df_sub = df.loc[:, ['season', 'casual', 'registered']].groupby('season').mean()
df_sub.plot.bar(stacked=True)
```
